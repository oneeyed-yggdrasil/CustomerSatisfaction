```python
# indentifying the categorical variables
cat_var= df.select_dtypes(include= ["string"]).columns.tolist()
cat_data = df.select_dtypes(include= ["string"])
print(cat_var)

for col in cat_var[1:]:
    print()
    print(f"\033[1m{col}\033[0m\n") #print column name above the plot
    ax = cat_data[col].value_counts().sort_index().plot(kind='bar', rot=0, xlabel=col,ylabel='count')
    abs_values = cat_data[col].value_counts(ascending=False).values

    ax.bar_label(container=ax.containers[0], labels=abs_values)
    plt.show()
```
